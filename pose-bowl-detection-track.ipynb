```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image

# Загрузка изображения
image = Image.open("/kaggle/input/posebowl/spaceship/images/test/00054819240f9d46378288b215dbcd3a.jpg")
# image = Image.open("/kaggle/working/images/000dbf763348037b46558bbcb6a032ac.png")
plt.imshow(image)

# Загрузка маски
with open("/kaggle/input/posebowl/spaceship/labels/test/00054819240f9d46378288b215dbcd3a.txt", "r") as file:
# with open("/kaggle/working/labels/000dbf763348037b46558bbcb6a032ac.txt", "r") as file:
    lines = file.readlines()

# Применение маски к изображению
for line in lines:
    class_id, x_center, y_center, width, height = map(float, line.split())
    print('coordinates:\n', x_center, y_center, width, height)
    left = (x_center - width / 2) * image.width
    top = (y_center - height / 2) * image.height
    rect_width = width * image.width
    rect_height = height * image.height
    rect = patches.Rectangle((left, top), rect_width, rect_height, linewidth=1, edgecolor='r', facecolor='none')
    plt.gca().add_patch(rect)

plt.show()
```

```python
def conver_to_yolo_format(values):
    x_min, y_min, x_max, y_max = values

    # Размеры изображения
    image_width = 1280
    image_height = 1024

    # Вычисляем центр ограничивающего прямоугольника
    x_center = (x_min + x_max) / 2
    y_center = (y_min + y_max) / 2

    # Вычисляем ширину и высоту ограничивающего прямоугольника
    width = x_max - x_min
    height = y_max - y_min

    # Нормализуем координаты и размеры
    x_center /= image_width
    y_center /= image_height
    width /= image_width
    height /= image_height
    return [x_center, y_center, width, height]
```
